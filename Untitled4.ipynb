```python
import numpy as np
import matplotlib.pyplot as plt
fm=1
t=np.arange(0,3,0.01)
a=np.sin(2*np.pi*fm*t)+1
plt.plot(t,a)
plt.legend()
```

**Sampling**

```python
fs=20*fm
ts=np.arange(0,3,(1/fs))
xs=np.sin(2*np.pi*fm*ts)+1
plt.plot(ts,xs,'o')
plt.plot(t,a,':')
```

**Quantization**

```python
n=3
L=2**n
x_min=np.min(xs)
x_max=np.max(xs)
delta=(x_max-x_min)/L
levels=np.arange(x_min,x_max+delta,delta)
level_values=[np.mean([levels[i],levels[i+1]]) for i in range(L)]


xsq=[]
for i in xs:
  p=int(np.floor((i-x_min)/delta))
  if p>7:
    p=7
  xsq.append(level_values[p]+x_min)

plt.plot(ts,xsq,'r*')
plt.hlines(levels,0,3,linestyle=':')
```

**Encoding**

```python
bin_list=np.array([int(np.floor((i-x_min)/delta)) for i in xs])
bin_list[bin_list==8]=7
print(bin_list)

pcm="".join([np.binary_repr(x,width=n) for x in bin_list])
print(pcm)
```

**SNR**

```python
bits=10
SNR=[]
SNR2=[]

for n in range(2,bits):
  L=2**n
  delta=(x_max-x_min)/L
  s_power=np.linalg.norm(a)**2/a.size
  n_power=delta**2/12
  SNR.append(10*np.log10(s_power/n_power))
  SNR2.append(1.8+6*n)
plt.plot(range(2,bits),SNR,'*:')
plt.figure()
plt.plot(range(2,bits),SNR2,'*:')
```

**BPSK**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bitgen(a):
  b=np.random.randint(0,2,a)
  return b

def carrier(phase):
  t=np.linspace(0,np.pi,100)
  c=np.cos((2*np.pi*t)+phase)
  e=np.linalg.norm(c)
  c=c/e
  return c

c=carrier(0)
print(sum(c**2))
plt.plot(carrier(0))
```

```python
def bpsk(x,c):
  l=np.array([])
  for bit in x:
    if bit==0:
      l=np.concatenate([l,-c])
    else:
      l=np.concatenate([l,c])

  return l
```

```python
def awgn(l,std):
  n=np.random.normal(0,std,l.size)
  l=l+n
  return l

def demod(l,c):
  con=np.split(l,len(l)/len(c))
  h=[]
  for i in con:
    h.append(np.sum(i*c))

  h=np.array(h)
  return h
```

```python
def ML_bpsk(a):
  a[a<=0]=0
  a[a>0]=1
  return a

def error(a,o):
  return np.average(np.abs(a-o))
```

```python
n=10
b=bitgen(n)
c=carrier(0)
l=bpsk(b,c)
i1=demod(l,c)
plt.plot(i1,np.zeros_like(i1),'o')
plt.vlines(0,-0.04,0.04,linestyle=':')
i11=ML_bpsk(i1)

print(b,i11)
print('error=',error(b,i11))
```

```python
l2=awgn(l,3)
i2=demod(l2,c)
print(i2)
plt.plot(i2,np.zeros_like(i2),'o')
plt.vlines(0,-0.04,0.04,linestyle=':')

i22=ML_bpsk(i2)

print(b,i22)
print('error=',error(b,i22))
```

```python
import math
from scipy.special import erfc
snr = np.arange(-4,13,2)
#snr per bit is taken as [-4,-2,0....12]
#snr=10*np.log10(Eb/No)
#Since carrier is normalized, Eb is taken as 1. Therefore from the equation, No = 1/(10**(snr/10))
No = 1/(10**(snr/10))
ber=np.zeros_like(No)
for i in range(len(No)):
    b=bitgen(1000)
    c=carrier(0)
    j=bpsk(b,c)
    j=awgn(j,math.sqrt(No[i]/2))
    j=demod(j,c)
    j=ML_bpsk(j)
    ber[i]=error(b,j)
BER_theory = 0.5*erfc(np.sqrt(10**(snr/10)))
print(ber)
plt.semilogy(snr,ber,'ro',label='BER')
plt.semilogy(snr,BER_theory,'*:',label='BER_theory')
plt.legend();
```

**SNR-BER**

```python
from scipy.special import erfc
import math

snr=np.arange(-4,12,2)
No=1/(10**(snr/10))
ber=np.zeros_like(No)
for i in range(len(No)):
  b=bitgen(1000)
  o=b[::2]
  e=b[1::2]
  c1=carrier(0)
  c2=carrier(np.pi/2)
  j1=bpsk(o,c1)
  j2=bpsk(e,c2)
  j=j1+j2
  j=awgn(j,math.sqrt(No[i]/2))
  p=demod(j,c1)
  q=demod(j,c2)
  p=ML_bpsk(p)
  q=ML_bpsk(q)
  X=[]
  for k in range(len(p)):
    X.append(p[k])
    X.append(q[k])

  ber[i]=error(X,b)

BER_theory = 0.25*erfc(np.sqrt(10**(snr/10)))
plt.semilogy(snr,ber,'o',label='BER')
plt.semilogy(snr,BER_theory,'*:',label='BER_theory')
plt.legend();
```
